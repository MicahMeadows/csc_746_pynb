```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
# b
dataset_a = np.genfromtxt('hw02_a.csv', delimiter=',')

# c 
x = dataset_a[:, 0]
y = dataset_a[:, 1]

# d 
x_max = np.max(x)
x_min = np.min(x)
print(f"Max value of x: {x_max}")
print(f"Min value of x: {x_min}")

# e 
y_mean = np.mean(y)
y_median = np.median(y)
print(f"Mean value of y: {y_mean}")
print(f"Median value of y: {y_median}")

# f 
high_profit = dataset_a[y > 300000]

# g 
plt.plot(x, y, 'bo')
plt.title("dataset a")
plt.xlabel("investment")
plt.ylabel("profit")
plt.show()
```
